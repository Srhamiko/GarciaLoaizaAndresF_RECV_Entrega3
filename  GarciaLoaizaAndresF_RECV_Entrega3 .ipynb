```python
class_0= y==(y[0])
class_1= y!=(y[0])
```

```python
plt.plot(x1[class_0],x2[class_0],'o', c='black', label='class 0')
plt.plot(x1[class_1],x2[class_1],'o', c='red', label='class 1')
#plt.xlim(0,1)
#plt.ylim(0,1)
plt.legend()
plt.show()
```

```python
#Funci√≥n que queremos encontrar
def y(x):
  return -(b+w_1*x)/w_2
#Funci√≥n de activaci√≥n
def act(z):
  return 1/(1+ np.exp(-z))
```

```python
data=pd.read_csv('iris.data')
#Definimos el perceptron

#Inicializamos los pesos en 0
b=0
w_1=0
w_2=0

#Voy a hacer que si la clase es igual a Iris-setosa sea igual a 1. sino es igual a 0 y remuevo la clase que sobra
IndiceABorrar=data[(data['clase']=="Iris-virginica")].index
data.drop(IndiceABorrar,inplace=True)

data.loc[data.clase =="Iris-setosa",'clase']=1
data.loc[data.clase =="Iris-versicolor",'clase']=0
#Inicializamos el proceso de aprendizaje y los n epocas
n_epochs=15
n=0.5
x1=data['slength']
x2=data['pwidth']
Y=data['clase']
np.unique(Y,return_counts=True)
#print(Y)
```

```python

for epoch in range(n_epochs):
  for i,j,k in zip(x1,x2,Y):
    z=i*w_1+j*w_2+b
    sig_z=act(z)
    if sig_z>=0.5:
      out=1
    if sig_z<0.5:
      out=0
    error =k-out
    b= b+n*error
    w_1=w_1+ n*error*i
    w_2=w_2+ n*error*j

  print('epoch [{}/{}], bias: {},w1: {},w2:{}'.format(epoch+1,n_epochs,b,w_1,w_2) ) 
  plt.plot(x1[class_0],x2[class_0],'o', c='black', label='class 0')
  plt.plot(x1[class_1],x2[class_1],'o', c='red', label='class 1')
  plt.plot(np.sort(x1),y(np.sort(x1)),c='g')
  #plt.xlim(0,1)
  #plt.ylim(0,1)
  plt.legend()
  plt.show()
```

# Segunda parte

```python
#Perceptron multicapas
data=pd.read_csv('iris.data')

X=data.drop(['clase'], axis='columns')
Y=data['clase']
Y=np.array(Y)
```

## Modelo de dos capas

```python
from sklearn.model_selection import train_test_split

X_train,X_test, Y_train, Y_test= train_test_split(X,Y)
```

```python
#Modelo 2 capas
from sklearn.neural_network import MLPClassifier

model= MLPClassifier(activation ='tanh', hidden_layer_sizes=(4,3),solver='adam',learning_rate_init=0.1)
model
```

```python
model.fit(X_train,Y_train)

Y_pred= model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_pred,Y_test )
```
